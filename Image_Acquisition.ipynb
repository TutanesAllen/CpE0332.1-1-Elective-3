```python
from google.colab import drive
drive.mount('/content/drive')
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
img = cv2.imread('/content/drive/MyDrive/Colab/orange.png')
```

```python
cv2_imshow(img)
```

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_img)
```

```python
img_flip = cv2.flip(img, 1)
cv2_imshow(img_flip)
```

```python
img_flip = cv2.flip(img, 0)
cv2_imshow(img_flip)
```

```python
img_rs = cv2.resize(img, (300, 300))
cv2_imshow(img_rs)
```

```python
img_rotate = cv2.rotate(img_rs, cv2.ROTATE_90_CLOCKWISE)
cv2_imshow(img_rotate)
```

```python
img_blur = cv2.blur(img, (30, 30))
cv2_imshow(img_blur)
```
